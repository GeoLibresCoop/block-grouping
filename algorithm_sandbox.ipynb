```python
# resumir en recorrer_vecindario que devuelve array de {'costo': costo, 'vecino': queda}
def vecindario(segmentacion):
    # devuelve array de {'costo': costo, 'vecino': queda}
    vecindario = []
    for s, segmento in enumerate(segmentacion):
        if len(segmento) == 1: # no se puede sacar
            pass
        else:
            if len(segmento) == 2: # se parte en 2
                resto = list(segmentacion); resto.pop(); 
                costo = costo_extraccion(0, segmento) + costo_segmentacion(resto)
                queda = [[segmento[0]], [segmento[1]]] + segmentacion[0:s] + segmentacion[s+1:]
                vecindario.extend([{'costo': costo, 'vecino': queda}])
            else: # se saca una manzana por vez
                for m, manzana in enumerate(segmento):
                    resto = list(segmentacion); resto.pop(s)
                    costo = costo_extraccion(m, segmento) + costo_segmentacion(resto)
                    queda = [[segmento[m]], segmento[0:m] + segmento[m+1:]] + segmentacion[0:s] + segmentacion[s+1:]
                    vecindario.extend([{'costo': costo, 'vecino': queda}])
            
    for o, origen in enumerate(segmentacion):
        for d, destino in enumerate(segmentacion[0:o] + segmentacion[o+1:]):        
            resto = list(segmentacion); resto.pop(o); resto.pop(d)
            if len(origen) > 1: # se puede sacar una manzana y queda resto
                for m, manzana in enumerate(origen):
                    costo = costo_transferencia(m, origen, destino) + costo_segmentacion(resto)
                    queda = [origen[0:m] + origen[m+1:], destino + [origen[m]]] + resto
                    vecindario.extend([{'costo': costo, 'vecino': queda}])
            else: # segmento unitario se fusiona origen con destino
                costo = costo_segmentacion([origen + destino]) + costo_segmentacion(resto)
                queda = [origen + destino] + resto
                vecindario.extend([{'costo': costo, 'vecino': queda}])
    return vecindario




segmentacion = [[0], [1, 2, 3]]
costo_actual = costo_segmentacion(segmentacion)
print 'segmentacion actual', segmentacion, 'costo: ', costo_actual 
print
vecindario = vecindario(segmentacion)
for vecino in vecindario:
    delta = vecino['costo'] - costo_actual
    #if delta < 0: 
    print delta, vecino
    
print "cantidad de vecinos:", len(vecindario)
```
